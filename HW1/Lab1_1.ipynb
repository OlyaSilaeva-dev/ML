5. Создайте случайный вектор размером 15 и замените максимальное значение на $NaN$.

7. Рассмотрим случайную матрицу  $15 × 2 $, представляющую декартовы координаты, преобразуйте их в полярные координаты.

```python
dec_coords = np.random.rand(15, 2)

r = np.sqrt(dec_coords[:, 0]**2 + dec_coords[:, 1]**2)
theta = np.arctan2(dec_coords[:, 1], dec_coords[:, 0])

polar_coords = np.column_stack((r, theta))
print(polar_coords)
```

8. По двум массивам $x$ и $y$ постройте матрицу Коши $C$
 $(C_{ij} =\frac{1}{x_i - y_j})$

```python
x = np.array([4, 5, 6])
y = np.array([1, 2, 1])

C = 1 / (x[:, np.newaxis] - y)
print(C)
```

9. Рассмотрим матрицу $10 × 3 $, извлеките строки с неравными значениями (например, $[2,2,3]$).

```python
arr = np.random.randint(1, 4, size=(10, 3))
print("Исходная матрица:")
print(arr)
filtered = arr[~np.all(arr == arr[:, [0]], axis=1)]
print("\nСтроки с неравными значениями:")
print(filtered)
```

10. Извлеките уникальные строки из двумерного массива.

# Найдите ответы на вопросы с помощью предложенного датасета:

```python
# Load dataset here
df = pd.read_csv('olympics_dataset.csv', delimiter=',')
```

1. Сколько спортсменов участвовали в Летних Олимпийских играх за всё время?

```python
df['Name'].nunique()
```

2. Найдите участника с наибольшим количеством медалей

```python
medals_df = df[df['Medal'] != 'No medal']
medal_counts = medals_df['Name'].value_counts()

top_athlete = medal_counts.idxmax()
top_medals = medal_counts.max()
print(f"Самый титулованный спортсмен: {top_athlete} с {top_medals} медалями")
```

3. Какой спорт является самым популярным, то есть в нём участвовало больше всего спортсменов за всё время?

```python
sport_counts = df.groupby('Sport')['Name'].nunique()
most_popular_sport = sport_counts.idxmax()
athlete_count = sport_counts.max()

print(f"Самый популярный спорт: {most_popular_sport} ({athlete_count} спортсменов)")
```

4. Кто был победителем медального зачёта в 2004 году?

```python
df_2004 = df[(df['Year'] == 2004) & (df['Medal'] != 'No nedal')]

medal_table = df_2004.pivot_table(index='NOC', columns='Medal', aggfunc='size', fill_value=0)
medal_table_sorted = medal_table.sort_values(by=['Gold', 'Silver', 'Bronze'], ascending=False)
top_country = medal_table_sorted.index[0]
print(f"Победитель медального зачёта в 2004 году: {top_country}")

top_gold = medal_table_sorted.loc[top_country, 'Gold']
top_silver = medal_table_sorted.loc[top_country, 'Silver']
top_bronze = medal_table_sorted.loc[top_country, 'Bronze']
print(f"Золото: {top_gold}, Серебро: {top_silver}, Бронза: {top_bronze}")
```

5. В каком виде спорта США завоевало наибольшее число медалей?

```python
usa_medals = df[(df['NOC'] == 'USA') & (df['Medal'] != 'No medal')]
sport_cnts = usa_medals['Sport'].value_counts()

top_sport = sport_cnts.idxmax()
top_count = sport_cnts.max()

print(f"США завоевали больше всего медалей в виде спорта: {top_sport} ({top_count} медалей)")
```

# С помощью графика покажите:

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python
summer_df = df[df["Season"] == "Summer"]
unique_athletes = summer_df.drop_duplicates(subset=["Year", "Name", "Sex"])

cnt_by_gender = unique_athletes.groupby(["Year", "Sex"]).size().unstack()
cnt_by_gender.plot(kind='bar', figsize=(14, 6), color=["lightpink", "skyblue"])
plt.title("Участие мужчин и женщин в Летних Олимпийских играх по годам")
plt.xlabel("Год")
plt.ylabel("Количество участников")
plt.legend(["Женщины", "Мужчины"])
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.tight_layout()
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
gold_medals = df[df['Medal'] == 'Gold']
gold_medals_cnt = gold_medals['Team'].value_counts()
top_10_countries = gold_medals_cnt.head(10)

plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='bar', color='gold', edgecolor='black')
plt.title("Топ-10 стран по количеству золотых медалей", fontsize=14)
plt.xlabel("Страны")
plt.ylabel("Количество золотых медалей")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(top_10_countries):
    plt.text(i, v + 0.05, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
ussr_df = df[(df['Team'] == 'Soviet Union') & (df['Medal'] != 'No medal')]

medal_colors = {
    'Gold': 'goldenrod',
    'Silver': 'silver',
    'Bronze': 'sienna',
}

medals_by_year = ussr_df.groupby(['Year', 'Medal']).size().unstack(fill_value=0)
medals_by_year.plot(kind='bar', stacked=True, color=[medal_colors[col] for col in medals_by_year.columns],figsize=(10, 6))
plt.title('Количество медалей СССР по типам по годам')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.xticks(rotation=45)
plt.legend(title='Медали', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
